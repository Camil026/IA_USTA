#Importando Librerias


```python
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import numpy as np
import matplotlib.pyplot as plt
```

#Ejercicio 1: Graficas Variable Temperatura

```python
temperatura=ctrl.Antecedent(np.arange(80,110,1),'temperatura')
temperatura['High']=fuzz.trapmf(temperatura.universe,[80,80,90,100])
temperatura['Very High']=fuzz.trapmf(temperatura.universe,[90,100,110,110])
temperatura.view()
```

#Ejercicio 2: Modificaci√≥n


```python
temperatura=ctrl.Antecedent(np.arange(60,110,1),'temperatura')
temperatura['Medium']=fuzz.trapmf(temperatura.universe,[60,60,70,80])
temperatura['High']=fuzz.trapmf(temperatura.universe,[70,80,90,100])
temperatura['Very High']=fuzz.trapmf(temperatura.universe,[90,100,110,110])
temperatura.view()
```
